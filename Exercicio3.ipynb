```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
hemograma = {
    'Leucócitos': [2000, 4000, 5000, 6500],
    'Plaquetas': [100000, 20000, 80000, 145000],
    'Linfócitos': [2.3, 4.5, 6, 5],
}


hemograma_df = pd.DataFrame(hemograma)


scaler = StandardScaler()
scaled_data = scaler.fit_transform(hemograma_df)


dbscan = DBSCAN(eps=0.5, min_samples=2)  
dbscan.fit(scaled_data)


labels = dbscan.labels_


hemograma_df['cluster'] = labels


n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print(f"Número de clusters encontrados: {n_clusters}")


plt.scatter(hemograma_df['Leucócitos'], hemograma_df['Plaquetas'], c=hemograma_df['cluster'])
plt.xlabel('Leucócitos')
plt.ylabel('Plaquetas')
plt.title('Clusters de Hemograma')
plt.show()
```
