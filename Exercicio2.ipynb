```python
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
```

```python
corrente_eletrica = np.array([5, 10, 14, 2, 1.5, 6])
tempo = np.array([1, 2, 4, 6, 7, 10])
```

```python
dados = np.column_stack((corrente_eletrica, tempo))
```

```python
modelo = IsolationForest(contamination=0.2)  
modelo.fit(dados)
```

```python
anomalias = modelo.predict(dados)
```

```python
amostras_anomalas = dados[anomalias == -1]
```

```python
plt.scatter(tempo, corrente_eletrica, label='Dados de Corrente')
plt.scatter(amostras_anomalas[:, 1], amostras_anomalas[:, 0], color='red', label='Anomalias')
plt.xlabel('Tempo')
plt.ylabel('Corrente Elétrica')
plt.legend()
plt.title('Detecção de Anomalias na Corrente Elétrica')
plt.show()
```
